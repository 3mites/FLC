```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Universal set, set of all values that moisture can become
moisture_values = np.linspace(0, 30, 100)
# Sensor Input
moisture =  18

# Membership functions, for low the range is 0-13%. for ideal 12-18%. for high 15-30%
low = fuzz.trimf(moisture_values, [0, 6.5, 12])
ideal = fuzz.trimf(moisture_values, [12, 13.5, 18])
high = fuzz.trimf(moisture_values, [15,23,100])

#scaling to make the membership degree of membership function(y-axis) range from 0-1
low /= np.max(low)
ideal /= np.max(ideal)
high /= np.max(high)

degree_low_moisture = fuzz.interp_membership(moisture_values, low, moisture)
degree_ideal_moisture = fuzz.interp_membership(moisture_values, ideal,moisture)
degree_high_moisture = fuzz.interp_membership(moisture_values, high, moisture)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(moisture_values, low, label='Low')
plt.plot(moisture_values, ideal, label='Ideal')
plt.plot(moisture_values, high, label='High')
plt.title('Membership Function for Moisture Content')
plt.xlabel('Moisture Content(%)')
plt.ylabel('Membership')
#plt.axvline(x=moisture, color='red', linestyle='--', label='Moisture Value')

plt.scatter(moisture,degree_low_moisture,s=degree_low_moisture*100, color='red', alpha=0.5)
plt.scatter(moisture,degree_ideal_moisture,s=degree_ideal_moisture*100, color='red', alpha=0.5)
plt.scatter(moisture,degree_high_moisture,s=degree_high_moisture*100, color='red', alpha=0.5)

plt.legend()
plt.savefig('moisture_membership_function.PNG')

print(f'Degree of Membership for {moisture}:')
print(f'- Low: {degree_low_moisture}')
print(f'- Ideal: {degree_ideal_moisture}')
print(f'- High: {degree_high_moisture}')
```
