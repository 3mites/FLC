```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl

# Universal Set for Temperature and Moisture
temperature = ctrl.Antecedent(np.arange(0, 101, 1), 'temperature')
moisture = ctrl.Antecedent(np.arange(0, 101, 1), 'moisture')

# Universal Set for Output; value of temperature adjustment
output_variable = ctrl.Consequent(np.arange(0, 101, 1), 'output_variable')

"""
Universal Set, for choosing values that fuzzy logic can pick to adjust temperature accordingly
#for low it will choose a value between 0 and 60 degrees celsius
#for high it will choose a value between 40 and 100 degrees celsius
# low = 0-41 degree celsius High = 40-100 degrees celsius""
output_variable['low'] = fuzz.trimf(output_variable.universe, [0, 20, 60])
output_variable['high'] = fuzz.trimf(output_variable.universe, [40, 80, 100])
"""
# Membership Functions 
temperature['low'] = fuzz.trimf(temperature.universe, [0, 20, 60])
temperature['high'] = fuzz.trimf(temperature.universe, [40, 80, 100])
moisture['low'] = fuzz.trimf(moisture.universe, [0, 6.5, 12])
moisture['ideal'] = fuzz.trimf(moisture.universe, [12, 13.5, 18])
moisture['high'] = fuzz.trimf(moisture.universe, [15, 23, 100])

output_variable['low'] = fuzz.trimf(output_variable.universe, [0, 20, 60])
output_variable['high'] = fuzz.trimf(output_variable.universe, [40, 80, 100])


"""
Defining rules:
if high temperature and moisture is in high then high temperature
if high temperature and moisture is in low then low temperature
if high temperature and moisture is in ideal then low temperature
if low  temperature and moisture is in high then high tempearture
if low  temperature and moisture is in low then low temperature
if low  temperature and moisture is in ideal then low temperature
"""
rule1 = ctrl.Rule(temperature['high'] & moisture['high'], output_variable['high'])
rule2 = ctrl.Rule(temperature['high'] & moisture['low'], output_variable['low'])
rule3 = ctrl.Rule(temperature['high'] & moisture['ideal'], output_variable['low'])
rule4 = ctrl.Rule(temperature['low'] & moisture['high'], output_variable['high'])
rule5 = ctrl.Rule(temperature['low'] & moisture['low'], output_variable['low'])
rule6 = ctrl.Rule(temperature['low'] & moisture['ideal'], output_variable['low'])
temperature_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])

# Deffuzification
temperature_sim = ctrl.ControlSystemSimulation(temperature_ctrl)
# Set input values
temperature_value = 27.60
moisture_value = 14
temperature_sim.input['temperature'] = temperature_value
temperature_sim.input['moisture'] = moisture_value
# Perform the simulation
temperature_sim.compute()

# Display the adjustment value
print(f"Temperature Adjustment: {temperature_sim.output['output_variable']:.2f}")

# Display the membership degrees
degree_low_temp = fuzz.interp_membership(temperature.universe, fuzz.trimf(temperature.universe, [0, 20, 60]), temperature_value)
degree_high_temp = fuzz.interp_membership(temperature.universe, fuzz.trimf(temperature.universe, [40, 80, 100]), temperature_value)

degree_low_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [0, 6.5, 12]), moisture_value)
degree_ideal_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [12, 13.5, 18]), moisture_value)
degree_high_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [15, 23, 100]), moisture_value)

# Display the plots
temperature.view(sim=temperature_sim)
moisture.view(sim=temperature_sim)
output_variable.view(sim=temperature_sim)

# Show the plots
plt.show()

```
