```python
%matplotlib inline
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl

"""
Set sensor input values will cause an error when moisture content is <13 since there is no rule for that
One thing I think of doing is terminating the process once the moisture values become <13.5%
Play with the values, when moisture value is above 18% fuzzy logic will suggest to increase the temperature
and as it decays it will suggest to lower temperatures
"""
temperature_value = 70
moisture_value = 14
humidity_value = 35

# Universal Set for Temperature and Moisture
temperature = ctrl.Antecedent(np.arange(0, 101, 1), 'temperature')
moisture = ctrl.Antecedent(np.arange(0, 31, 1), 'moisture')
humidity = ctrl.Antecedent(np.arange(0,91,1), 'humidity')


# Universal Set for Output; value of temperature adjustment
output_variable = ctrl.Consequent(np.arange(0, 101, 1), 'output_variable')

"""
Universal Set, for choosing values that fuzzy logic can pick to adjust temperature accordingly
#for low it will choose a value between 0 and 60 degrees celsius
#for high it will choose a value between 40 and 100 degrees celsius
# low = 0-41 degree celsius High = 40-100 degrees celsius""
output_variable['low'] = fuzz.trimf(output_variable.universe, [0, 20, 60])
output_variable['high'] = fuzz.trimf(output_variable.universe, [40, 80, 100])
"""
# Membership Functions 
temperature['low'] = fuzz.trimf(temperature.universe, [0, 20, 50])
temperature['warm'] = fuzz.trimf(temperature.universe, [40,50,60])
temperature['high'] = fuzz.trimf(temperature.universe, [55, 82, 100])

humidity['high'] = fuzz.trimf(humidity.universe, [40,65,90])
humidity['low'] = fuzz.trimf(humidity.universe, [20,40,60])
humidity['very low'] = fuzz.trimf(humidity.universe, [0,20,40])

moisture['ideal'] = fuzz.trimf(moisture.universe,[12,13,15])
moisture['high'] = fuzz.trimf(moisture.universe, [14, 22, 30])

output_variable['low'] = fuzz.trimf(output_variable.universe, [0, 20, 50])
output_variable['warm'] = fuzz.trimf(output_variable.universe, [40,50,60])
output_variable['high'] = fuzz.trimf(output_variable.universe, [55, 82, 100])

rule1_t = ctrl.Rule((temperature['high']|temperature['warm']|temperature['low']) & 
                    (moisture['high']|moisture['ideal']) & 
                     humidity['high'], 
                    output_variable['low'])


rule2_t = ctrl.Rule(temperature['high'] & moisture['high'] & (humidity['low']|humidity['very low']),
                     output_variable['high'])


rule3_t = ctrl.Rule(temperature['high'] & moisture['ideal'] & (humidity['low']|humidity['very low']),
                     output_variable['warm'])


rule4_t = ctrl.Rule(temperature['warm'] & moisture['high'] & (humidity['low']|humidity['very low']),
                     output_variable['high'])

rule5_t = ctrl.Rule(temperature['warm'] & moisture['ideal'] & (humidity['low']|humidity['very low']),
                  output_variable['warm'])


rule6_t = ctrl.Rule(temperature['low'] & moisture['high'] & (humidity['low']|humidity['very low']),
                  output_variable['high'])


rule7_t = ctrl.Rule(temperature['low'] & moisture['ideal'] & (humidity['low']|humidity['very low']),
                  output_variable['warm'])

rule8_t = ctrl.Rule((temperature['low']|temperature['warm']|temperature['high']) & 
                    moisture['ideal'] & 
                    (humidity['low']|humidity['very low']|humidity['high']),
                  output_variable['low'])



#temperature_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
temperature_ctrl = ctrl.ControlSystem([rule1_t, rule2_t, rule3_t, rule4_t, rule5_t,rule6_t,rule7_t,rule8_t])



# Deffuzification
temperature_sim = ctrl.ControlSystemSimulation(temperature_ctrl)


# Perform the simulation
temperature_sim.input['temperature'] = temperature_value
temperature_sim.input['moisture'] = moisture_value
temperature_sim.input['humidity'] = humidity_value
temperature_sim.compute()



# Display the adjustment value
print(f"Moisture Value: {moisture_value}")
print(f"Temperature Value: {temperature_value}")
print(f"Humidity Value: {humidity_value}")
print(f"Temperature Adjustment: {temperature_sim.output['output_variable']:.2f}")

# Display the membership degrees
#degree_low_temp = fuzz.interp_membership(temperature.universe, fuzz.trimf(temperature.universe, [0, 20, 60]), temperature_value)
#degree_high_temp = fuzz.interp_membership(temperature.universe, fuzz.trimf(temperature.universe, [40, 80, 100]), temperature_value)

#degree_low_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [0, 6.5, 12]), moisture_value)
#degree_ideal_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [12, 13.5, 18]), moisture_value)
#degree_high_moisture = fuzz.interp_membership(moisture.universe, fuzz.trimf(moisture.universe, [15, 23, 100]), moisture_value)

# Display the plots
temperature.view(sim=temperature_sim)
moisture.view(sim=temperature_sim)
humidity.view(sim=temperature_sim)
output_variable.view(sim=temperature_sim)
# Show the plots
plt.show()

```
